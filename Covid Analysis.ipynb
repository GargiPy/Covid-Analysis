```python
to_edit = data[['location', 'population', 'aged_65_older_percent', 'gdp_per_capita', 'hospital_beds_per_thousand']]
to_edit = to_edit.drop_duplicates()
to_edit
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Assuming you have a DataFrame named 'df' with columns: 'feature1', 'feature2', 'target'

# Split the DataFrame into two sets: one with missing values and one without missing values
df_missing = to_edit[to_edit['hospital_beds_per_thousand'].isnull()]
df_not_missing = to_edit[~to_edit['hospital_beds_per_thousand'].isnull()]

# Separate the features and target variables
X_train = df_not_missing[['population']]
y_train = df_not_missing['hospital_beds_per_thousand']

# Create a Random Forest regressor
rf_model = RandomForestRegressor()

# Fit the model on the non-missing data
rf_model.fit(X_train, y_train)

# Predict the missing values using the trained model
X_test = df_missing[['population']]
predicted_values = rf_model.predict(X_test)

# Fill the missing values in the 'target' column with the predicted values
to_edit.loc[to_edit['hospital_beds_per_thousand'].isnull(), 'hospital_beds_per_thousand'] = predicted_values

# Print the updated DataFrame with the filled missing values
to_edit.info()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Assuming you have a DataFrame named 'df' with columns: 'gdp_per_capita', 'feature1', 'feature2', ...

# Split the data into training and test sets
training_data = to_edit[to_edit['gdp_per_capita'].notnull()]
test_data = to_edit[to_edit['gdp_per_capita'].isnull()]

# Select the features and target variable for training
features = ['population', 'hospital_beds_per_thousand']
target = 'gdp_per_capita'

X_train = training_data[features]
y_train = training_data[target]

# Train the decision tree model
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

# Use the trained model to predict the missing values
X_test = test_data[features]
predicted_values = model.predict(X_test)

# Assign the predicted values to the missing values in the original DataFrame
to_edit.loc[to_edit['gdp_per_capita'].isnull(), 'gdp_per_capita'] = predicted_values

# Print the DataFrame with the missing values filled
to_edit.info()
```

```python
from sklearn.mixture import GaussianMixture

# Using Gaussian Misture Model
training_data = to_edit[to_edit['aged_65_older_percent'].notnull()]
test_data = to_edit[to_edit['aged_65_older_percent'].isnull()]

# Select the features and target variables
features = ['population', 'hospital_beds_per_thousand', 'gdp_per_capita']

# Prepare the training set by extracting the features
X_train = training_data[features]

# Train the GMM on the training set
gmm = GaussianMixture(n_components=3)  # You can adjust the number of components as needed
gmm.fit(X_train)

# Predict the missing values using the trained GMM
X_test = test_data[features]
predicted_values = gmm.predict(X_test)

# Assign the predicted values to the missing values
to_edit.loc[to_edit['aged_65_older_percent'].isnull(), 'aged_65_older_percent'] = predicted_values

# Check the dataset
to_edit.info()
```
